### About Dataset


A non-banking financial institution (NBFI) or non-bank financial company (NBFC) is a Financial Institution that does not have a full banking license or is not supervised by a national or international banking regulatory agency. NBFC facilitates bank-related financial services, such as investment, risk pooling, contractual savings, and market brokering.

An NBFI is struggling to mark profits due to an increase in defaults in the vehicle loan category. The company aims to determine the client’s loan repayment abilities and understand the relative importance of each parameter contributing to a borrower’s ability to repay the loan.

```python
# Mengubah tipe data menjadi string
str_convert = ['Client_Gender', 
               'Client_Education', 
               'Client_Marital_Status', 
               'Client_Occupation', 
               'Client_Income_Type', 
               'Client_Housing_Type', 
               'Loan_Contract_Type']

for col in str_convert:
    df[col] = df[col].astype(str)
```

### Distrbusi Nasabah Gagal Bayar

```python
# Membuat 2 subplots
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,5))
# Mengatur warna pada plot
palette = sns.color_palette("Set2")
sns.set_palette(palette)

# subplot 1: bar plot
default_type = sns.countplot(x='Default', data=df, ax=ax1)
# Menambahkan keterangan di atas bar
for p in default_type.patches:
    ax1.annotate(int(p.get_height()),
                 (p.get_x() + p.get_width()/2., p.get_height()),
                 ha='center',
                 va='center',
                 textcoords='offset points',
                 xytext=(0,5)
                 )
# Mengatur judul dan label
ax1.set_title("Distribusi Nasabah Gagal Bayar",
          fontweight= 'bold',
          fontsize = 13)
ax1.set_xlabel("Default",
           fontweight= 'bold')
ax1.set_ylabel("Jumlah Nasabah",
           fontweight= 'bold')
# Mengatur xticks
ax1.set_xticks(ticks=[0,1], 
           labels=["Tidak Gagal Bayar", "Gagal Bayar"])


# subplot 2: pie chart
default_type_count = df['Default'].value_counts()
# Membuat label untuk pie chart
pie_label = {0: "Tidak Gagal Bayar", 1: "Gagal Bayar"}
# Mengatur warna pada pie chart
pie_colors = [palette[0], palette[1]] 
# Membuat pie chart
ax2.pie(
    default_type_count,
    labels=[pie_label[i] for i in default_type_count.index],
    autopct='%1.1f%%',
    startangle=90,
    wedgeprops={'linewidth':5, 'edgecolor': 'white'}
)
centre_circle = plt.Circle((0,0), 0.4, fc='white')
ax2.add_artist(centre_circle)

# Mengatur judul 
ax2.set_title('Persentase Nasabah Gagal Bayar',
              fontweight= 'bold',
              fontsize = 13)

# Menampilkan plot
plt.show()
```

Dari visualisasi tersebut dapat diketahui bahwa proporsi nasabah yang membayar pinjaman dengan lancar lebih besar dibanding nasabah yang gagal membayar pinjaman. Adapun persentase nasabah tidak gagal bayar sebesar 93% dan nasabah gagal bayar sebesar 7%. Ini dapat menjadi indikator yang serius karena jika nilai nasabah gagal bayar meningkat melebihi 10%, Bank perlu mengembangkan sistem penilaian risiko yang lebih ketat sebelum menyetujui pinjaman nasabah.

### Hubungan Tingkat Pendidikan Nasabah dengan Risiko Gagal Bayar

```python
# Mengatur ukuran visualisasi plot
plt.figure(figsize=(10,5))
# Membuat barplot
ax = sns.barplot(x='Client_Education', y='Default', data=df, color='mediumaquamarine')

# Menambah keterangan di atas bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2,
            height+0.01,
            f'{height*100:.1f}%',
            ha='center'
            )

# Mengatur judul dan label
plt.title("Hubungan Tingkat Pendidikan dengan Risiko Gagal Bayar", fontweight='bold', fontsize=13)
plt.xlabel("Tingkat Pendidikan", fontweight='bold')
plt.ylabel("Rata-rata Default", fontweight='bold')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

Dari visualisasi tersebut dapat diketahui rata-rata proporsi nasabah yang mengalami gagal bayar dalam setiap kelompok pendidikan. Kesimpulan yang dapat ditarik adalah semakin tinggi tingkat pendidikan nasabah, maka semakin rendah risiko gagal bayar. Untuk itu, tingkat pendidikan dapat digunakan sebagai salah satu pertimbangan utama dalam model credit scoring, nasabah dengan tingkat pendidikan Graduation dan post Grad bisa dianggap low-risk, sedangkan kelompok nasabah yang berpendidikan rendah dapat diberikan pemahaman serta penilaian tambahan atau batas kredit yang disesuaikan. Selain itu, Bank juga dapat menawarkan program literasi keuangan khusus untuk kelompok nasabah yang berpendidikan menengah untuk meningkatkan kemampuan pengelolaan kredit.
